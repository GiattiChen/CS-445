## Part 1 Toy Problem (20 pts)

```python
def toy_reconstruct(toy_img):
    """
    The implementation for gradient domain processing is not complicated, but it is easy to make a mistake, so let's start with a toy example. Reconstruct this image from its gradient values, plus one pixel intensity. Denote the intensity of the source image at (x, y) as s(x,y) and the value to solve for as v(x,y). For each pixel, then, we have two objectives:
    1. minimize (v(x+1,y)-v(x,y) - (s(x+1,y)-s(x,y)))^2
    2. minimize (v(x,y+1)-v(x,y) - (s(x,y+1)-s(x,y)))^2
    Note that these could be solved while adding any constant value to v, so we will add one more objective:
    3. minimize (v(1,1)-s(1,1))^2
    
    :param toy_img: numpy.ndarray
    """
    v = np.zeros((toy_img.shape[0], toy_img.shape[1], toy_img.shape[2]))
    for i in range(toy_img.shape[2]):
        im = toy_img[:,:,i]
        im_h, im_w = im.shape
        im2var = np.arange(im_h*im_w).reshape(im_w, im_h).T
        count = 0
        A = np.zeros((im_h*(im_w-1) + (im_h-1)*im_w + 1, im_h*im_w))
        b = np.zeros(A.shape[0])
        e = 0
        for y in range(im_h):
            for x in range(im_w-1):
                # print(e, im2var[y][x+1], 1)
                # print(e, im2var[y][x], -1)
                A[e][im2var[y][x+1]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = im[y][x+1] - im[y][x] 
                e += 1
        for y in range(im_h-1):
            for x in range(im_w):
                # print(e, im2var[y+1][x], 1)
                # print(e, im2var[y][x], -1)
                A[e][im2var[y+1][x]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = im[y+1][x] - im[y][x] 
                e += 1
        A[e][im2var[0][0]] = 1 
        A_sparse = csr_matrix(lil_matrix(A))
        b[e] = im[0][0] 
        v[:,:,i] = scipy.sparse.linalg.lsqr(A_sparse, b)[0].reshape(im_w, im_h).T
    return np.abs(v)
            
    pass
```

## Part 2 Poisson Blending (50 pts)

```python
import scipy
from scipy.sparse import *
def poisson_blend(cropped_object, object_mask, background_img):
    """
    :param cropped_object: numpy.ndarray One you get from align_source
    :param object_mask: numpy.ndarray One you get from align_source
    :param background_img: numpy.ndarray 
    """
    #TO DO 
    # select four boundary points and minimize (v(x,y)-s(x,y))^2 
    import pdb
    col, row = np.where(object_mask==True)
    min_col_arr = np.where(col==np.min(col))[0]
    min_col_sample = min_col_arr[0]
    max_col_arr = np.where(col==np.max(col))[0]
    max_col_sample = max_col_arr[0]
    min_row_arr = np.where(row==np.min(row))[0]
    min_row_sample = min_row_arr[0]
    max_row_arr = np.where(row==np.max(row))[0]
    max_row_sample = max_row_arr[0]
    
    # for the pixels in the mask region, do toy_reconstruct
    v = background_img[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,:]*0
    out = cropped_object[:,:,:]
    for i in range(background_img.shape[2]):
        im = cropped_object[col[min_col_sample] : col[max_col_sample]+1, row[min_row_sample] : row[max_row_sample] + 1, i]
        S = background_img[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,i]
        im_h, im_w = im.shape
        im2var = np.arange(im_h*im_w).reshape(im_w, im_h).T
        count = 0
        A = np.zeros((im_h*(im_w-1) + (im_h-1)*im_w + 4, im_h*im_w))
        b = np.zeros(A.shape[0])
        e = 0
        for y in range(im_h):
            for x in range(im_w-1):
                A[e][im2var[y][x+1]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = im[y][x+1] - im[y][x] 
                e += 1
        for y in range(im_h-1):
            for x in range(im_w):
                A[e][im2var[y+1][x]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = im[y+1][x] - im[y][x] 
                e += 1
        A[e][im2var[0][row[min_col_sample]-row[min_row_sample]]] = 1 
        b[e] = S[0][row[min_col_sample]-row[min_row_sample]] 
        e+=1
        A[e][im2var[-1][row[max_col_sample]-row[min_row_sample]]] = 1 
        b[e] = S[-1][row[max_col_sample]-row[min_row_sample]] 
        e+=1
        A[e][im2var[col[min_row_sample]-col[min_col_sample]][0]] = 1 
        b[e] = S[col[min_row_sample]-col[min_col_sample]][0] 
        e+=1
        A[e][im2var[col[max_row_sample]-col[min_col_sample]][-1]] = 1 
        b[e] = S[col[max_row_sample]-col[min_col_sample]][-1] 
        e+=1
        A_sparse = csr_matrix(lil_matrix(A))
        v[:,:,i] = scipy.sparse.linalg.lsqr(A_sparse, b)[0].reshape(im_w, im_h).T
    out[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,:] = v
    return  np.abs(out*np.stack((object_mask, object_mask, object_mask),axis=2)+background_img*(1-np.stack((object_mask, object_mask, object_mask),axis=2)))
    pass
```

## Part 3 Mixed Gradients (20 pts)

```python
def mix_blend(cropped_object, object_mask, background_img):
    """
    :param cropped_object: numpy.ndarray One you get from align_source
    :param object_mask: numpy.ndarray One you get from align_source
    :param background_img: numpy.ndarray 
    """
    #TO DO 
    col, row = np.where(object_mask==True)
    min_col_arr = np.where(col==np.min(col))[0]
    min_col_sample = min_col_arr[0]
    max_col_arr = np.where(col==np.max(col))[0]
    max_col_sample = max_col_arr[0]
    min_row_arr = np.where(row==np.min(row))[0]
    min_row_sample = min_row_arr[0]
    max_row_arr = np.where(row==np.max(row))[0]
    max_row_sample = max_row_arr[0]
    
    # for the pixels in the mask region, do toy_reconstruct
    v = background_img[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,:]*0
    out = cropped_object[:,:,:]
    for i in range(background_img.shape[2]):
        im = cropped_object[col[min_col_sample] : col[max_col_sample]+1, row[min_row_sample] : row[max_row_sample] + 1, i]
        S = background_img[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,i]
        im_h, im_w = im.shape
        im2var = np.arange(im_h*im_w).reshape(im_w, im_h).T
        count = 0
        A = np.zeros((im_h*(im_w-1) + (im_h-1)*im_w + 4, im_h*im_w))
        b = np.zeros(A.shape[0])
        e = 0
        for y in range(im_h):
            for x in range(im_w-1):
                A[e][im2var[y][x+1]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = compare(im[y][x+1] - im[y][x], S[y][x+1] - S[y][x])
                e += 1
        for y in range(im_h-1):
            for x in range(im_w):
                A[e][im2var[y+1][x]] = 1
                A[e][im2var[y][x]] = -1
                b[e] = compare(im[y+1][x] - im[y][x], S[y+1][x] - S[y][x])
                e += 1
        A[e][im2var[0][row[min_col_sample]-row[min_row_sample]]] = 1 
        b[e] = S[0][row[min_col_sample]-row[min_row_sample]] 
        e+=1
        A[e][im2var[-1][row[max_col_sample]-row[min_row_sample]]] = 1 
        b[e] = S[-1][row[max_col_sample]-row[min_row_sample]] 
        e+=1
        A[e][im2var[col[min_row_sample]-col[min_col_sample]][0]] = 1 
        b[e] = S[col[min_row_sample]-col[min_col_sample]][0] 
        e+=1
        A[e][im2var[col[max_row_sample]-col[min_col_sample]][-1]] = 1 
        b[e] = S[col[max_row_sample]-col[min_col_sample]][-1] 
        e+=1
        A_sparse = csr_matrix(lil_matrix(A))
        v[:,:,i] = scipy.sparse.linalg.lsqr(A_sparse, b)[0].reshape(im_w, im_h).T
    out[col[min_col_sample]:col[max_col_sample]+1,row[min_row_sample]:row[max_row_sample]+1,:] = v
    return  np.abs(out*np.stack((object_mask, object_mask, object_mask),axis=2)+background_img*(1-np.stack((object_mask, object_mask, object_mask),axis=2)))
    pass

def compare(a, b):
    if abs(a) > abs(b):
        return a
    else:
        return b
```

## Color2Gray (20 pts)

```python
def color2gray(img):
    v = np.zeros((img.shape[0], img.shape[1]))
    im = img[:,:,:]
    im_h, im_w, im_d = im.shape
    im2var = np.arange(im_h*im_w).reshape(im_w, im_h).T
    count = 0
    A = np.zeros((im_h*(im_w-1) + (im_h-1)*im_w + min(im_h, im_w), im_h*im_w))
    b = np.zeros(A.shape[0])
    e = 0
    for y in range(im_h):
        for x in range(im_w-1):
            A[e][im2var[y][x+1]] = 1
            A[e][im2var[y][x]] = -1
            b[e] = compare_three(im[y][x+1][0] - im[y][x][0], im[y][x+1][1] - im[y][x][1], im[y][x+1][2] - im[y][x][2])
            e += 1
    for y in range(im_h-1):
        for x in range(im_w):
            A[e][im2var[y+1][x]] = 1
            A[e][im2var[y][x]] = -1
            b[e] = compare_three(im[y+1][x][0] - im[y][x][0], im[y+1][x][1] - im[y][x][1], im[y+1][x][2] - im[y][x][2])
            e += 1
    for i in range(0, min(im_h, im_w), 2):
        A[e][im2var[i][i]] = 1 
        b[e] = compare_three(im[i][i][0], im[i][i][1], im[i][i][2])
        e += 1
    A_sparse = csr_matrix(lil_matrix(A))
    
    v[:,:] = scipy.sparse.linalg.lsqr(A_sparse, b)[0].reshape(im_w, im_h).T
    return v
    pass

def compare_three(a, b, c):
    d = {}
    d[abs(a)] = a
    d[abs(b)] = b
    d[abs(c)] = c
    return d[max(abs(a), abs(b), abs(c))]
```

## Laplacian pyramid blending (20 pts)

```python
import pdb
import scipy
def laplacian_blend(img1, img2):
    out = np.zeros(img1.shape)
    sigma = 5
    kernel_half_size = 4* sigma
    gaussian_k = utils_hybrid.gaussian_kernel(kernel_half_size, sigma)
    impulse_k = scipy.signal.unit_impulse(gaussian_k.shape, 'mid')
    lap_pyramid1 = []
    lap_pyramid2 = []
    cur_img1 = img1[:,:,:]
    cur_img2 = img2[:,:,:]
    for i in range(3):
        gau1_im = np.zeros(cur_img1.shape)
        gau2_im = np.zeros(cur_img2.shape)
        for j in range(img1.shape[2]):
            gau1_im[:,:,j] = filter_image(cur_img1[:,:,j], gaussian_k)
            gau2_im[:,:,j] = filter_image(cur_img2[:,:,j], gaussian_k)
        lap_pyramid1.append(cur_img1 - gau1_im)
        lap_pyramid2.append(cur_img2 - gau2_im)
        # plt.figure()
        # plt.imshow(np.abs(lap_pyramid1[-1]))
        
        new_img1 = cur_img1[::2,::2,:]
        new_img2 = cur_img2[::2,::2,:]
        for j in range(img1.shape[2]):
            new_img1[:,:,j] = filter_image(cur_img1[:,:,j], gaussian_k)[::2,::2]
            new_img2[:,:,j] = filter_image(cur_img2[:,:,j], gaussian_k)[::2,::2]
        
        if i != 2:
            cur_img1 = new_img1
            cur_img2 = new_img2
            
    out = np.zeros(cur_img1.shape)
    for power in range(2, -1, -1):
        # mask
        mask_left = np.ones((out.shape[0], out.shape[1]))
        mask_right = np.ones((out.shape[0], out.shape[1]))
        sigma = 2**(4-power)
        something, mask = np.mgrid[0:1:1/out.shape[0], 0:1:sigma/(0.25*out.shape[1])]
        mask_left[:,int(out.shape[1]/2 - out.shape[1]/(8*sigma)):int(out.shape[1]/2 - out.shape[1]/(8*sigma)) + mask.shape[1]] = 1 - mask
        mask_left[:,int(out.shape[1]/2 + out.shape[1]/(8*sigma)):] = 0
        mask_right[:,int(out.shape[1]/2 - out.shape[1]/(8*sigma)):int(out.shape[1]/2 - out.shape[1]/(8*sigma)) + mask.shape[1]] = mask
        mask_right[:,:int(out.shape[1]/2 - out.shape[1]/(8*sigma))] = 0
        # kernel
        kernel_half_size = 4*sigma
        gaussian_k = utils_hybrid.gaussian_kernel(kernel_half_size, sigma)
        impulse_k = scipy.signal.unit_impulse(gaussian_k.shape, 'mid')
        lap_k = impulse_k - gaussian_k
        # filt the image
        # filter_image_out = np.zeros(img1.shape)
        for i in range(img1.shape[2]):
            #im_left = filter_image(img1[:,:,i] ,lap_k)
            #im_right = filter_image(img2[:,:,i] ,lap_k)
            #plt.figure()
            #plt.imshow(im_left*mask_left)
            #plt.figure()
            #plt.imshow(im_right*mask_right)
            if power == 2:
                out[:,:,i] = cur_img1[:,:,i]*mask_left + cur_img2[:,:,i]*mask_right
            out[:,:,i] += lap_pyramid1[power][:,:,i]*mask_left + lap_pyramid2[power][:,:,i]*mask_right
        #plt.figure()
        #plt.imshow(out)
        if power != 0:
            out = upsample_2(out)[:lap_pyramid1[power-1].shape[0], :lap_pyramid1[power-1].shape[1]]
    return out
    pass


def upsample_2(im):
    old_im = np.zeros((im.shape[0]+1, im.shape[1]+1, im.shape[2]))
    old_im[:-1,:-1,:] = im
    old_im[-1,:-1,:] = im[-1,:,:]
    old_im[:-1,-1,:] = im[:,-1,:]
    old_im[-1,-1,:] = (old_im[-1,-2,:] + old_im[-2,-1,:])/2
    new_im = np.zeros((im.shape[0]*2, im.shape[1]*2, im.shape[2]))
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            new_im[2*i, 2*j, :] = old_im[i, j, :]
            new_im[2*i+1, 2*j, :] = (old_im[i+1, j, :] + old_im[i, j, :])/2
            new_im[2*i, 2*j+1, :] = (old_im[i, j, :] + old_im[i, j+1, :])/2
            new_im[2*i+1, 2*j+1, :] = (old_im[i, j, :] + old_im[i+1, j+1, :])/2
    return new_im
def filter_image(im, fil): 
    '''
    im: H x W floating point numpy ndarray representing image in grayscale 
    fil: M x M floating point numpy ndarray representing 2D filter
    '''
    H, W = im.shape
    hs = fil.shape[0] // 2 # half of filter size
    fftsize = 4096   # should be order of 2 (for speed) and include padding
    im_fft = np.fft.fft2(im, (fftsize, fftsize)) 
    fil_fft = np.fft.fft2(fil, (fftsize, fftsize)) 
    im_fil_fft = im_fft * fil_fft;
    im_fil = np.fft.ifft2(im_fil_fft)
    im_fil = im_fil[hs:hs + H, hs:hs + W] 
    im_fil = np.real(im_fil)
    return im_fil
```
